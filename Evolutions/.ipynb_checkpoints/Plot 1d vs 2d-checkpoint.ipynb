```python
import sys
sys.path.append(r"C:\Users\asant\Documents\GitHub\Tensor-Network")

import dmrg2 as dmrg
import SpinSystems_2d as SS2d
import numpy as np
import matplotlib.pyplot as plt
from ncon import ncon
import MPO_class as MPO
import MPS_class as MPS
import seaborn as sns
import tdvp2 as tdvp
import matplotlib as mpl
#%matplotlib notebook 

import h5py
```

```python
file_list = []
psi_list = []
MzMpo_list = []
L_list = []

for L in range(3,10):
    file_list.append(h5py.File(f"{L}x{L}_Floquet_eps0.1_chi64.h5","r"))
    psi_list.append(MPS.MPS(L*L,2,2))
    MzMpo_list.append( MPO.getMzMPO(L*L) ) 
    L_list.append(L)

n_max_list = []
for i in range(len(file_list)):
    n_max_list.append(max(list(map(int,file_list[i]['MPS/n'].keys()))) )

Mz_tot = []
err_   = []
SurvProb = []
for i in range(len(file_list)):
    Mz_tot.append( np.zeros(n_max_list[i]) )
    err_.append( np.zeros(n_max_list[i]) )
    SurvProb.append( np.zeros(n_max_list[i]) )
    
mzloc = [[] for i in range(len(L_list))]


for j, file in enumerate(file_list):
    for n in range(n_max_list[j]):
        psi_list[j].load_hdf5(file,n)
        Mz_tot[j][n] = MzMpo_list[j].contractMPOMPS(psi_list[j]).real/L_list[j]**2
        
        mzloc[j].append( MPO.return_LocalMz(psi_list[j]).real.reshape(L_list[j],L_list[j]) )
        
        mzj = mzloc[j][-1]
        err_sym = np.max([np.abs(mzj-mzj[:,::-1]).ravel(),\
                  np.abs(mzj-mzj[::-1,:]).ravel(),\
                  np.abs(mzj-mzj.T).ravel()])
        
        err_[j][n] = err_sym
        
        # Compute SurvProb
        sigma = n%2
        M_list = psi_list[j].M
        sp = M_list[0][:,sigma,:]
        for i in range(1,len(M_list)):
            sp = ncon([sp,M_list[i][:,sigma,:]],[[-1,1],[1,-2]])
        SurvProb[j][n] = np.abs(sp[0][0])**2
```

```python
file_list = []
psi_list = []
MzMpo_list = []
L_list = []

for L in range(3,10):
    file_list.append(h5py.File(f"1d_{L}x{L}_Floquet_eps0.1_chi64.h5","r"))
    psi_list.append(MPS.MPS(L*L,2,2))
    MzMpo_list.append( MPO.getMzMPO(L*L) ) 
    L_list.append(L)

n_max_list = []
for i in range(len(file_list)):
    n_max_list.append(max(list(map(int,file_list[i]['MPS/n'].keys()))) )

Mz_tot_1d = []
err_1d   = []
SurvProb_1d = []

for i in range(len(file_list)):
    Mz_tot_1d.append( np.zeros(n_max_list[i]) )
    err_1d.append( np.zeros(n_max_list[i]) )
    SurvProb_1d.append( np.zeros(n_max_list[i]) )

mzloc_1d = [[] for i in range(len(L_list))]

for j, file in enumerate(file_list):
    for n in range(n_max_list[j]):
        
        psi_list[j].load_hdf5(file,n)
        Mz_tot_1d[j][n] = MzMpo_list[j].contractMPOMPS(psi_list[j]).real/L_list[j]**2
        
        mzloc_1d[j].append( MPO.return_LocalMz(psi_list[j]).real.reshape(L_list[j],L_list[j]) )
        
        mzj = mzloc[j][-1]
        err_sym = np.max([np.abs(mzj-mzj[:,::-1]).ravel(),\
                  np.abs(mzj-mzj[::-1,:]).ravel(),\
                  np.abs(mzj-mzj.T).ravel()])
        
        err_1d[j][n] = err_sym
        
        
        # Compute SurvProb
        sigma = n%2
        M_list = psi_list[j].M
        sp = M_list[0][:,sigma,:]
        for i in range(1,len(M_list)):
            sp = ncon([sp,M_list[i][:,sigma,:]],[[-1,1],[1,-2]])
        SurvProb_1d[j][n] = np.abs(sp[0][0])**2
```

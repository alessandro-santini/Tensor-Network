```python
import sys
sys.path.append(r"C:\Users\asant\Documents\GitHub\Tensor-Network")

import dmrg2 as dmrg
import SpinSystems_2d as SS2d
import numpy as np
import matplotlib.pyplot as plt
from ncon import ncon
import MPO_class as MPO
import MPS_class as MPS
import seaborn as sns
import tdvp2 as tdvp
import matplotlib as mpl
#%matplotlib notebook 

import h5py

file_MPS = h5py.File(f"TFI_11x11_Evolution_MPS_h0.1.h5")
dt = 0.5
```

```python
L = 11
psit =  MPS.MPS(L*L,2,2)
MzMPO = MPO.getMzMPO(L*L)
#%%%%%%%%%%%%%%%%%%%%%%        
n_max   = max(list(map(int,file_MPS['MPS/n'].keys())))
mz      = np.zeros((n_max))
err_    = np.zeros_like(mz)
deltaE  = np.zeros_like(mz)
mz_loc  = np.zeros((n_max,11,11))
ent_ent = np.zeros((n_max,120))

J = 1
h = 0.1
H = SS2d.IsingMPO_2D(L,L,h,J)

psit.load_hdf5(file_MPS,0)
E0 = H.contractMPOMPS(psit).real

for n in range(n_max):
    psit.load_hdf5(file_MPS,n)
    
    mz[n] = MzMPO.contractMPOMPS(psit).real/121
    mz_loc[n,:,:] = MPO.return_LocalMz(psit).real.reshape(L,L).real

    mzj = mz_loc[n,:,:].real
    err_sym = np.max([np.abs(mzj-mzj[:,::-1]).ravel(),\
              np.abs(mzj-mzj[::-1,:]).ravel(),\
              np.abs(mzj-mzj.T).ravel()])
    err_[n] = err_sym
    
    deltaE[n] = (H.contractMPOMPS(psit).real-E0)/121
    
    ent_ent[n,:] = psit.compute_EntEntropy().real
```

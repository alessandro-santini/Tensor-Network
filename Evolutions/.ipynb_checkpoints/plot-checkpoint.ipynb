```python
import sys
sys.path.append(r"C:\Users\asant\Documents\GitHub\Tensor-Network")

import dmrg2 as dmrg
import SpinSystems_2d as SS2d
import numpy as np
import matplotlib.pyplot as plt
from ncon import ncon
import MPO_class as MPO
import MPS_class as MPS
import seaborn as sns
import tdvp2 as tdvp
import matplotlib as mpl
#%matplotlib notebook 

import h5py
```

```python
file_list = []
psi_list = []
MzMpo_list = []
L_list = []

for L in range(3,9):
    file_list.append(h5py.File(f"{L}x{L}_Floquet_eps0.1_chi64.h5","r"))
    psi_list.append(MPS.MPS(L*L,2,2))
    MzMpo_list.append( MPO.getMzMPO(L*L) ) 
    L_list.append(L)
        
n_max = max(list(map(int,file_list[0]['MPS/n'].keys())))
mz    = np.zeros((n_max,len(file_list)))
err_  = np.zeros_like(mz)

mzloc = [[] for i in range(len(L_list))]
for n in range(n_max):
    for j, file in enumerate(file_list):
        psi_list[j].load_hdf5(file,n)
        mz[n,j] = MzMpo_list[j].contractMPOMPS(psi_list[j]).real/L_list[j]**2
        
        mzloc[j].append( MPO.return_LocalMz(psi_list[j]).real.reshape(L_list[j],L_list[j]) )
        
        mzj = mzloc[j][-1]
        err_sym = np.max([np.abs(mzj-mzj[:,::-1]).ravel(),\
                  np.abs(mzj-mzj[::-1,:]).ravel(),\
                  np.abs(mzj-mzj.T).ravel()])
        
        err_[n,j] = err_sym
```

```python
L = 9

file_list.append(h5py.File(f"{L}x{L}_Floquet_eps0.1_chi64.h5","r"))
psi_list.append(MPS.MPS(L*L,2,2))
MzMpo_list.append( MPO.getMzMPO(L*L) ) 
L_list.append(L)

n_max_9 = max(list(map(int,file_list[-1]['MPS/n'].keys())))
mz_9    = np.zeros(n_max_9)
err_9   = np.zeros(n_max_9)

mzloc_9 = []
j = -1
for n in range(n_max_9):
    psi_list[j].load_hdf5(file_list[-1],n)
    mz_9[n] = MzMpo_list[j].contractMPOMPS(psi_list[j]).real/L_list[j]**2

    mzloc_9.append( MPO.return_LocalMz(psi_list[j]).real.reshape(L_list[j],L_list[j]) )

    mzj = mzloc_9[-1]
    err_sym = np.max([np.abs(mzj-mzj[:,::-1]).ravel(),\
              np.abs(mzj-mzj[::-1,:]).ravel(),\
              np.abs(mzj-mzj.T).ravel()])

    err_9[n] = err_sym
```

```python
plt.rc('text',usetex=True)
plt.figure(figsize=(12,7))
for i in range(6):
    plt.plot(np.abs(mz[:,i]),'-o',markersize=4,label=r'$N = %d$'%(L_list[i]**2))
plt.plot(np.abs(mz_9[:]),'-o',markersize=4,label=r'$N = %d$'%(L_list[-1]**2))    
plt.tick_params(labelsize=22)
plt.xlabel("$n$",fontsize=32)
plt.ylabel(r"$|\langle \sigma_z\rangle|$",fontsize=32)
plt.legend(fontsize=20,ncol=4)

plt.ylim([0.,1.05])
#plt.savefig("magnetisation.pdf",format='pdf',bbox_inches='tight')
```

```python
plt.rc('text',usetex=True)
plt.figure(figsize=(12,7))
for i in range(6):
    plt.plot(err_[:,i],'-o',markersize=4,label=r'$N = %d$'%(L_list[i]**2))
plt.plot(err_9[:],'-o',markersize=4,label=r'$N = %d$'%(L_list[-1]**2))

plt.tick_params(labelsize=22)

plt.xlabel("$n$",fontsize=32)
plt.ylabel(r"$\Delta I$",fontsize=32)
plt.legend(fontsize=20,ncol=4)

plt.yscale('log')
plt.ylim([1e-6,1e-1])

#plt.savefig("error.pdf",format='pdf',bbox_inches='tight')
```

```python
file_long = h5py.File("long_6x6_Floquet_eps0.1_chi64.h5",'r')
L_ = 6
psi_long = MPS.MPS(L_*L_,2,2)
n_long  = max(list(map(int,file_long['MPS/n'].keys())))

Mz_MPO = MPO.getMzMPO(L_*L_)
mz_long = []
err_long = []
mz_loc_long = []

H = SS2d.IsingMPO_2D(6,6,0.,1.)
en = []
ent_ent = np.zeros((n_long,35))
for n in range(n_long):
    psi_long.load_hdf5(file_long,n)
    mz_long.append(Mz_MPO.contractMPOMPS(psi_long).real/(L_**2))
    
    mzj = MPO.return_LocalMz(psi_long).reshape(6,6)
    err_sym = np.max([np.abs(mzj-mzj[:,::-1]).ravel(),\
                  np.abs(mzj-mzj[::-1,:]).ravel(),\
                  np.abs(mzj-mzj.T).ravel()])
    err_long.append(err_sym)   
    mz_loc_long.append(mzj)
    en.append(H.contractMPOMPS(psi_long)/(L_**2))
    ent_ent[n,:] = psi_long.compute_EntEntropy().real
```

```python
import sys
sys.path.append(r"C:\Users\asant\Documents\GitHub\Tensor-Network")

import dmrg2 as dmrg
import SpinSystems_2d as SS2d
import numpy as np
import matplotlib.pyplot as plt
from ncon import ncon
import MPO_class as MPO
import MPS_class as MPS
import seaborn as sns
import tdvp2 as tdvp
import matplotlib as mpl
#%matplotlib notebook 

import h5py
```

```python
L = 81
file_01 = h5py.File(f"Evolutions/1d_9x9_Floquet_eps0.1_chi64.h5","r")
file_03 = h5py.File(f"Evolutions/1d_9x9_Floquet_eps0.3_chi64.h5","r")

data01  = np.loadtxt("Evolutions/theory/data_0.1.dat")
data03  = np.loadtxt("Evolutions/theory/data_0.3.dat")
```

```python
n_max  = max(list(map(int,file_01['MPS/n'].keys()))) 

psi01  = MPS.MPS(81,2,2)
psi03  = MPS.MPS(81,2,2)

Mz01 = np.zeros(n_max)
Mz03 = np.zeros(n_max)

MzMPO = MPO.getMzMPO(L)
```

```python
for n in range(n_max):
    psi01.load_hdf5(file_01,n)
    psi03.load_hdf5(file_03,n)
    
    Mz01[n] = MzMPO.contractMPOMPS(psi01)/L
    Mz03[n] = MzMPO.contractMPOMPS(psi03)/L
```

```python
colors = sns.color_palette()
plt.rc('text',usetex=True)
t_list = np.arange(0,n_max)
plt.plot(t_list,np.abs(Mz01),'^',markersize=3,color=colors[0],label=r"$\varepsilon=0.1$")
plt.plot(t_list,np.abs(Mz03),'^',markersize=3,color=colors[1],label=r"$\varepsilon=0.3$")

plt.plot(data01[:,1],data01[:,0],color=colors[0],label=r"$\varepsilon=0.1$")
plt.plot(data03[:,1],data03[:,0],color=colors[1],label=r"$\varepsilon=0.3$")

plt.legend()
```

```python
import dmrg2 as dmrg
import SpinSystems_1d as SS1d
import numpy as np
import matplotlib.pyplot as plt
from ncon import ncon
import MPO_class as MPO
import MPS_class as MPS
import seaborn as sns
import tdvp2 as tdvp
import matplotlib as mpl
```

```python
L = 121
h = 0.1
J = 1
H = SS1d.IsingMPO(L,h,J)
```

```python
psi0  = MPS.getAllUp(L)
delta = 4

for n in range(L//2-delta,L//2+delta+1):
    print(n)
    psi0.M[n] = ncon([psi0.M[n],np.array([[0,1],[1,0]])],[[-1,1,-3],[1,-2]])

time_engine = tdvp.TDVP2(psi0,H)
```

```python
plt.plot(MPO.return_LocalMz(psi0).real)
```

```python
mz_list = []
t_list = []
Mz_list = []
ent_ent = []

MzMPO = MPO.getMzMPO(L)

dt = 0.5
k  = 5
t_list.append(0.)

mz_list.append(MPO.return_LocalMz(time_engine.MPS).real)
Mz_list.append(MzMPO.contractMPOMPS(time_engine.MPS).real)
ent_ent.append(time_engine.MPS.compute_EntEntropy())

E0 = H.contractMPOMPS(time_engine.MPS)
time_engine.initialize()
for j in range(101):
    print(j,H.contractMPOMPS(time_engine.MPS)-E0)
    
    # Evolve the state of t1->t2, t2-t1=0.5
    for _ in range(k):
        time_engine.time_step(dt/k,1e-12)#1e-10)
        
    if time_engine.end_max:
        break
    t_list.append((j+1)*dt)
    mz_list.append(MPO.return_LocalMz(time_engine.MPS).real)
    Mz_list.append(MzMPO.contractMPOMPS(time_engine.MPS).real)
    ent_ent.append(time_engine.MPS.compute_EntEntropy())
    
```

```python
mz = np.array(mz_list)
```

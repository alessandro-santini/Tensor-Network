```python
import dmrg2 as dmrg
import SpinSystems_2d as SS2d
import numpy as np
import matplotlib.pyplot as plt
from ncon import ncon
import MPO_class as MPO
import MPS_class as MPS
import seaborn as sns
import tdvp2 as tdvp
import matplotlib as mpl
#%matplotlib notebook 

import h5py
file_MPS_25 = h5py.File("15x15_Evolution_MPS_h0.25.h5","r")
file_MPS_1  = h5py.File("15x15_Evolution_MPS_h0.1.h5","r")
```

```python
def get_index(i,j,Ly):
    return i+j*Ly
```

```python
Lx,Ly = 15,15
dt = 0.5

L=Lx*Ly

psi0 = MPS.getAllUp(Lx*Ly)

delta_x = 2
delta_y = 2
for i in range(Lx//2-delta_x,Lx//2+delta_x+1):
    for j in range(Ly//2-delta_y,Ly//2+delta_y+1):
        n = get_index(i,j,Ly)
        psi0.M[n] = ncon([psi0.M[n],np.array([[0,1],[1,0]])],[[-1,1,-3],[1,-2]])

psit_1 = MPS.getAllUp(Lx*Ly)
psit_25 = MPS.getAllUp(Lx*Ly)
```

```python
n_max_1 = max(list(map(int,file_MPS_1['MPS/n'].keys())))
n_max_25 = max(list(map(int,file_MPS_25['MPS/n'].keys())))
```

```python
MzMPO = MPO.getMzMPO(L)

mz1 = []
mz1.append(MPO.return_LocalMz(psi0).real)
ent_ent1 = []
Mz1 = []

mz25 = []
mz25.append(MPO.return_LocalMz(psi0).real)
ent_ent25 = []
Mz25 = []

for n in range(n_max_1):
    psit_1.load_hdf5(file_MPS_1,n)
    mz1.append(MPO.return_LocalMz(psit_1).real)
    ent_ent1.append(psit_1.compute_EntEntropy())
    Mz1.append(MzMPO.contractMPOMPS(psit_1).real/L)
    if n < n_max_25:
        psit_25.load_hdf5(file_MPS_25,n)
        mz25.append(MPO.return_LocalMz(psit_25).real)
        ent_ent25.append(psit_25.compute_EntEntropy())
        Mz25.append(MzMPO.contractMPOMPS(psit_1).real/L)
mz1  = np.array(mz1)
mz25 = np.array(mz25)
```

```python
mz1 = mz1.reshape((n_max_1+1,Lx,Ly))
mz25 = mz25.reshape((n_max_25+1,Lx,Ly))
```
